```python
model = pe.ConcreteModel()
model.I = pe.Set(initialize=I)
model.J = pe.Set(initialize=J)
model.P = pe.Set(initialize=P)
model.K = pe.Set(initialize=K)
```

```python
model.c_ijk = pe.Param(model.I, model.J, model.K, initialize={
  (i, j, k): C_ijk.get((i, j, k), 10000000) for i in model.I for j in model.J for k in model.K[i]
})


#model.c_ijk = pe.Param(model.I, model.J, model.K, initialize=C_ijk, default= 10000000)
model.c_i = pe.Param(model.I, initialize=C_i, default=0)
model.f_ik = pe.Param(model.I, model.K, initialize={
    (i, k): F_ik.get((i, k), 10000000) for i in model.I for k in model.K[i]
})
#model.f_ik = pe.Param(model.I, model.K, initialize=F_ik, default= 10000000)
model.d_jp = pe.Param(model.J, model.P, initialize=D_jp, default=0)
model.x_k_min = pe.Param(model.K, initialize={
    k: X_k_min.get(k, 0) for i in model.I for k in model.K[i]
})
#model.x_k_min = pe.Param(model.K, initialize=X_k_min, default= 0)
model.x_k_max = pe.Param(model.K, initialize={
    k: X_k_max.get(k, 0) for i in model.I for k in model.K[i]
})
#model.x_k_max = pe.Param(model.K, initialize=X_k_max, default=0)
model.X_p = pe.Param(model.P, initialize=X_p, default= 0)
model.q_i = pe.Param(model.I, initialize=Q_i, default= 0)
model.s_ij = pe.Param(model.I, model.J, initialize=S_ij, default= 0)
```

```python
model.x_ijpk = pe.Var(model.I, model.J, model.P, model.K, domain=pe.Reals, bounds=(0, 10000000))
model.y_ijk = pe.Var(model.I, model.J, model.K, domain=pe.Binary)
model.y_jik = pe.Var(model.J, model.I, model.K, domain=pe.Binary)
model.z_ik = pe.Var(model.I, model.K, domain=pe.Binary)
```

```python
obj_expr = sum(model.c_ijk[i,j,k] * sum(model.x_ijpk[i,j,p,k] for p in model.P) for i in model.I for j in model.J for k in model.K[i]) + sum(model.f_ik[i,k]*model.z_ik[i,k] for i in model.I for k in model.K[i] ) + sum(model.c_i[i] * sum(model.x_ijpk[i,j,p,k] for j in model.J for p in model.P for k in model.K[i]) for i in model.I )
model.objective = pe.Objective(sense=pe.minimize, expr=obj_expr)
```

```python
model.Constraint2 = pe.ConstraintList()
for j in model.J:
    for p in model.P:
        rhs = model.d_jp[j,p]
        lhs = sum(model.s_ij[i,j]*model.x_ijpk[i,j,p,k] for i in model.I for k in model.K[i])
        model.Constraint2.add(lhs == rhs)
```

```python
model.Constraint3 = pe.ConstraintList()
for i in model.I:
    for j in model.J:
        for p in model.P:
            for k in model.K[i]:
                rhs = model.x_ijpk[i,j,p,k]
                lhs = 0
                model.Constraint3.add(lhs <= rhs)

model.Constraint4 = pe.ConstraintList()
for i in model.I:
    for j in model.J:
        for p in model.P:
            for k in model.K[i]:
                lhs = model.x_ijpk[i,j,p,k]
                rhs = model.d_jp[j,p]
                model.Constraint4.add(lhs <= rhs)
```

```python
model.Transportation_ijpkConstraint3 = pe.ConstraintList()
for i in model.I:
    for k in model.K[i]:
        cen = sum(model.x_ijpk[i,j,p,k] for j in model.J for p in model.P)
        lhs = model.x_k_min[k]*model.z_ik[i,k]
        model.Transportation_ijpkConstraint3.add(lhs <= cen)

model.Transportation_ijpkConstraint4 = pe.ConstraintList()
for i in model.I:
    for k in model.K[i]:
        cen = sum(model.x_ijpk[i,j,p,k] for j in model.J for p in model.P)
        rhs = model.x_k_max[k]*model.z_ik[i,k]
        model.Transportation_ijpkConstraint4.add(cen <= rhs)
```

```python
model.Supply_ipConstraint4 = pe.ConstraintList()
for i in model.I:
    rhs = model.q_i[i]
    lhs = sum(sum(model.x_ijpk[i,j,p,k] for j in model.J for k in model.K[i])/model.X_p[p] for p in model.P)
    model.Supply_ipConstraint4.add( lhs<= rhs)
```

```python
solver = po.SolverFactory('gurobi')
```

```python
results = solver.solve(model, tee=True)
```
